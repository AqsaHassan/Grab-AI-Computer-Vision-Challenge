### Sharpen images

```python
import os
from PIL import Image
from PIL import ImageFilter
```

```python
filelist = os.listdir('data/train')
len(filelist)
```

#### Apply sharpen filter to train images

```python
count = 1
filesize = len(os.listdir('data/train'))
print(filesize)
for filelistcount in range(filesize):
    file_dir = os.path.join('data/train',os.listdir('data/train')[filelistcount])
    
    filesize = len(os.listdir(file_dir))
    print(filesize) #can see each folder image size

    item = os.listdir(file_dir)

    for imagecount in range(filesize):
        imagefile = item[imagecount]
        print(imagefile)
        im=Image.open(os.path.join(file_dir, item[imagecount]))
        im=im.convert("RGB")
        
        #Apply Sharpen filter
        im_sharp = im.filter(ImageFilter.SHARPEN)
        im_sharp.save(os.path.join(file_dir, str(count)+'sh_'+imagefile))
```
